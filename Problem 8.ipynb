```python
# Import Libraries

import string
```

```python
# Positive and Negative Word Dictionaries

positive_words = {
    "good", "great", "excellent", "amazing",
    "awesome", "happy", "love", "best",
    "fantastic", "nice", "wonderful"
}

negative_words = {
    "bad", "worst", "terrible", "awful",
    "hate", "poor", "sad", "horrible",
    "disappointing", "ugly", "boring"
}
```

```python
# Stopword List

stopwords = {
    "is", "the", "and", "a", "an",
    "of", "to", "in", "it", "this",
    "that", "was", "for", "on"
}
```

```python
# Sentiment Analysis Function

def analyze_sentiment(text):
    
    text = text.lower()
    
    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )
    
    words = [
        word for word in text.split()
        if word not in stopwords
    ]
    
    positive_count = 0
    negative_count = 0
    
    for word in words:
        
        if word in positive_words:
            positive_count += 1
        
        elif word in negative_words:
            negative_count += 1
    
    if positive_count > negative_count:
        return "Positive"
    
    elif negative_count > positive_count:
        return "Negative"
    
    return "Neutral"
```

```python
# User Input

user_text = input("Enter Text: ")

sentiment = analyze_sentiment(user_text)

print("Predicted Sentiment:", sentiment)
```
